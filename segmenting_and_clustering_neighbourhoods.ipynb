#importing necessary packages


```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from geopy.geocoders import Nominatim 
from pandas.io.json import json_normalize  
import folium 
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

Now we would scraping dataset for toronto

```python
source = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=954731857').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
df2.drop(df2.columns[[3]], axis = 1, inplace = True)
df2
```

```python
df2.shape
```

Now importing the CSV file containing latitudes and longitudes for various neighbourhoods in Canada

```python
coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
coordinates.head()
```

now merging the two pandas dataframe

```python
coordinates.rename(columns={'Postal Code':'Postcode'},inplace=True)
coordinates
```

```python
df3 = pd.merge(df2,coordinates,on = coordinates['Postcode'])
df3.drop(df3.columns[[0,4]],axis = 1,inplace = True)
```

```python
df3.head()
```

Now getting all rows in Borough column which have "toronto" as city

```python
y = []
temp = 0
for i in df3["Borough"]:
   
    if "Toronto" in i:
        y.append(df3.loc[temp])
    temp += 1
df4 = pd.DataFrame(y)
df4   
```

Now mapping all neighbourhoods in folium map

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

Now clustering the neighbourhoods using k-means clustering


```python
df5 = df4.drop(['Postal code','Borough',"Neighborhood"], axis=1)
df5.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = df5.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
```

```python
clusterNum = 5
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df4.insert(0, 'Cluster Labels', labels)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(clusterNum)
ys = [i + x + (i*x)**2 for i in range(clusterNum)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
